```python
# import statements
import pandas as pd
import sys
sys.path.append("../src")

from ncbi_upload import merge_metadata, generate_metadata_status, generate_alternative_id, convert_metadata, recover_vaccine_date
from gsheet_interact import gisaid_interactor
```

```python
# load the latest metadata and merge
repo_metadata_path = "/Users/karthikramesh/src/HCoV-19-Genomics/metadata.csv"
online_metadata_config = "../bjorn.ini"
merged_combined_metadata = merge_metadata(repo_metadata_path, online_metadata_config)
merged_combined_metadata
```

```python
ucsd_samples = pd.read_csv("../data/2021-10-29_04-46-10-all.bjorn_summary.csv")
ww_samples = ucsd_samples[ucsd_samples["Specimen source"] == "wastewater"]["SEARCH SampleID"].to_list()
len(ww_samples)
```

```python
new_id_data = generate_alternative_id(merged_combined_metadata)
new_id_data
```

```python
new_id_data = new_id_data[(~new_id_data["ID"].isin(ww_samples)) | (~new_id_data["alternative_ID"].isin(ww_samples))].reset_index(drop=True)
```

```python
# load and convert the metadata for samples that are not from wastewater
non_wastewater_converted_data = convert_metadata(new_id_data,
                                                 "../metadata_column_mapping.json",
                                                 "../metadata_constant_mapping.json",
                                                 "../author_conversions.csv",
                                                False)
non_wastewater_converted_data
```

```python
need_to_upload_non_ww = non_wastewater_converted_data[(~non_wastewater_converted_data["collected_by"].str.contains("Helix")) & (~non_wastewater_converted_data["gisaid_accession"].isna())]
need_to_upload_non_ww
```

```python
need_to_upload_non_ww = recover_vaccine_date(need_to_upload_non_ww).drop(columns=["last_vaccinated_raw", "date_of_sars_cov_2_vaccination"])
need_to_upload_non_ww
```

```python
# final cleanup
need_to_upload_non_ww[["collection_date", "geo_loc_name", "isolate", "isolation_source"]] = need_to_upload_non_ww[["collection_date", "geo_loc_name", "isolate", "isolation_source"]].fillna("not collected")
final_file = need_to_upload_non_ww.drop_duplicates(subset=["sample_name"]).fillna("")
final_file["collection_date"] = pd.to_datetime(final_file["collection_date"]).dt.strftime('%Y-%m-%d')
final_file["vaccine_received"].replace({"vaccine": "", "dose": ""}, inplace=True)
final_file["host"] = final_file["host"].str.replace("Human", "Homo Sapiens")
final_file = final_file[(final_file["host"] == "Homo Sapiens") | (final_file["host"] == "")]
final_file["host"] = "Homo Sapiens"
final_file["isolate"] = final_file["isolate"].str.replace("N/A", "not collected")
final_file["isolation_source"] = final_file["isolate"].str.replace("N/A", "not collected")
final_file
```

```python
final_file.to_csv("../data/ncbi_upload_v2/upload_ready_metadata.csv", index=False)
```
